```python
df = df.dropna()

# sanity checks
df = df[df['Acquisition_Cost'] > 0]
df = df[df['Conversion_Rate'] >= 0]

# after cleaning the dataset

df.to_csv("marketing_campaign_cleaned.csv", index=False)
```

```python
#how much it cost get one conversion
df['Cost_per_Conversion'] = df['Acquisition_Cost'] / (df['Conversion_Rate'] + 0.0001)
df.to_csv("marketing_campaign_cleaned.csv", index=False)
df.head()

```

```python
df['Profit_Score'] = df['ROI'] - df['Acquisition_Cost']
df.head()

```

```python
def strategy(roi):
    if roi > 0.5:
        return "Scale"
    elif roi > 0:
        return "Optimize"
    else:
        return "Stop"

df['Strategy'] = df['ROI'].apply(strategy)
df.head()
```

```python
#6. Exploratory Data Analysis (EDA)
#6.1 ROI Distribution

import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['ROI'], bins=30)
plt.title("ROI D#istribution")
plt.show()
```

```python
#6.2 Channel Performance
channel_perf = df.groupby('Channel_Used').agg({
    'ROI':'mean',
    'Conversion_Rate':'mean',
    'Acquisition_Cost':'mean'
}).reset_index()

channel_perf.sort_values(by='ROI', ascending=False)
```

```python
#6.3 Campaign Type Analysis
df.groupby('Campaign_Type')['ROI'].mean().sort_values()
```

```python
#6.4 Audience Analysis
df.groupby('Target_Audience')['Conversion_Rate'].mean().sort_values()
```

```python
#7.1 Channel + Audience (HIGH VALUE)
pivot = df.pivot_table(
    values='ROI',
    index='Channel_Used',
    columns='Target_Audience',
    aggfunc='mean'
)

sns.heatmap(pivot, annot=True, cmap='coolwarm')
plt.show()
```

```python
#7.2 Identify Bad Campaigns (CRITICAL)
bad_campaigns = df[df['ROI'] < 0]
bad_campaigns.shape
```

```python
#7.3 High Cost + Low ROI
high_cost_low_roi = df[
    (df['Acquisition_Cost'] > df['Acquisition_Cost'].mean()) &
    (df['ROI'] < df['ROI'].mean())
]
df.head()
```
